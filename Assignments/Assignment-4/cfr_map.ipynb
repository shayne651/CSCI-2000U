```python
#Shayne Taylor
#100552047
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import csv
%pylab inline

#reading the data from the CSV file
with open('Source_data_for_CFR_vaccine_map_abridged.csv') as csvfile:
    readCSV=csv.reader(csvfile,delimiter=',')
    lat=[]
    long= []
    category=[]
    cases=[]
    
    for row in readCSV:
        category.append(row[0])
        lat.append(float(row[1]))
        long.append(float(row[2]))
        cases.append(float(row[3]))
        
#making all the numpy arrays
npLat=np.array(lat)
npLong=np.array(long)
npCat=np.array(category)
npCases=np.array(cases)
```

```python
#drawing the map
m = Basemap(projection='mill')
m.drawcoastlines()

#logical indexing 
RLat=npLat[npCat == "Rubella"]
RLon=npLong[npCat == "Rubella"]
RCases=npCases[npCat == "Rubella"]

#Finds the maximum amount of cases
max=0
for p in range(len(RCases)):
    if (RCases[p]>max):
        max=RCases[p]

#plots the points with size depending how many cases there were
for p in range(len(RCases)):
    if (RCases[p]==max):
        x,y=m(RLon[p],RLat[p])
        m.plot(x,y,'mo',markersize=50,alpha=.5)
    elif (RCases[p]>4000):
        size=RCases[p]/max
        x,y=m(RLon[p],RLat[p])
        m.plot(x,y,'mo',markersize=(50*size),alpha=0.5)
    else:
        x,y=m(RLon[p],RLat[p])
        m.plot(x,y,'mo',markersize=(5),alpha=0.5)
#i didnt put in any safties to make sure each dot is atleast size 5 in this one because I calculated what they would need to be 

plt.title('Rubella outbreaks, 100552047')
plt.show()
```
